```python
import datetime as dt
df_age['birthday'] = pd.to_datetime(df_age['birthday'], errors='coerce')
```

```python
df_age['age'] = (dt.datetime.today() - df_age['birthday']) / (np.timedelta64(1, 'Y'))
df_age['age'] = df_age['age'].round()
```

```python
df_age = df_age[df_age['age'] > 0]
```

```python
# Наблюдаемое отклонение от среднего
dev = (df_age['age']-df_age['age'].mean()).abs()
# Стандартное отклонение
std = df_age['age'].std()
# Фильтруем исходный набор данных
df_age_new = df_age[~(dev>3*std)]
```

```python
df_start = df_prir.groupby(['title', 'student_id']).agg({'start_date': 'first'}).reset_index()
```

```python
### YOUR CODE HERE ###
df_start['start_date'] = pd.to_datetime(df_start['start_date'], errors='coerce')
df_start['Month'] = df_start['start_date'].dt.to_period('M')
```

```python
df_start = df_start.groupby(['title', 'Month'])['student_id'].count().reset_index()
```

```python
df_start['total'] = df_start.groupby('title')['student_id'].cumsum()
```

```python
df_start['Month'] = df_start['Month'].astype(str)
```

```python
df_start
```

```python
pivoted = pd.pivot_table(df_start, index='Month', columns='title', values='total')
pivoted = pivoted.fillna(0)
pivoted
```

```python
fig, axes = plt.subplots(5, 3, figsize=(20,50))
b = 0

for row in range(5):
    for column in range(3):
        i = pivoted.columns[b]
        axes[row, column].plot(pivoted[i].index, pivoted[i].values ,label='{}'.format(i), color='b')
        axes[row, column].set_title('{}'.format(i))
        axes[row, column].set_xlabel('Дата')
        axes[row, column].set_xticklabels(pivoted[i].index,rotation=90)
        axes[row, column].set_ylabel('Прирост')
        axes[row, column].grid()
        axes[row, column].legend(loc='upper left', title='Курс')
        b += 1
```
